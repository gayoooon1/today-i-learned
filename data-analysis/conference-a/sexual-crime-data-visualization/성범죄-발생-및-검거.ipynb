```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
import platform

from matplotlib import font_manager, rc 
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system')
```

```python
df = pd.read_csv('C:\\Users\\Owner\\Documents\\통신매체이용음란_발생_및_검거.csv', encoding='utf-8', index_col=0 )
#print(df)

df2 = df.dropna()
df2 = df2.set_index('항목.1')
df2 = df2[['2014년','2015년','2016년','2017년','2018년']]
df2['2014년'] = df2['2014년'].str.replace(',','')
df2['2015년'] = df2['2015년'].str.replace(',','')
df2['2016년'] = df2['2016년'].str.replace(',','')
df2['2017년'] = df2['2017년'].str.replace(',','')
df2['2018년'] = df2['2018년'].str.replace(',','')

df3 = df2.astype(float)
df3 = df3.T
df3
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
sns.lineplot(x = df3.index , y='발생건수', data = df3)
sns.lineplot(x = df3.index , y='검거건수', data = df3)
plt.title('통신메체 이용 음란 발생 및 검거건수')
plt.ylabel('발생건수 및 검거건수')
plt.savefig('통신메체 이용 음란 발생 및 검거건수.png')
```
