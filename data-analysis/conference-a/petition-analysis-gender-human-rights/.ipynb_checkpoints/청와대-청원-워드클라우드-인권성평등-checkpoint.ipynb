### 데이터 모으기   
본 데이터프레임은 청원시작일 기준 17.8 ~ 20.8의 청원 목록을 담고 있으며, 참여인원이 10000명이 넘는 목록만 수집했음을 밝힙니다.

```python
import pandas as pd

df_old = pd.read_csv('C:\\Users\\Owner\\Documents\\카카오톡 받은 파일\\final-blue-house-petition-best.csv', index_col = 0)
```

```python
import re 
import time
from tqdm import tqdm_notebook # PYTHON 진행표시바 만들기
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
pet_title = []
pet_start = []
pet_end = []
pet_ppl = []
pet_url = []
category = []

content = []



#def get_data() :
for pgn in range(1,6):
    urlbd = 'https://www1.president.go.kr/petitions/best?page='
    pg = urlopen(urlbd + str(pgn))
    soup = BeautifulSoup(pg, "html.parser")

    titles = soup.select('.b_list.category > .bl_body li .bl_wrap > .bl_subject > a.cb')
    urls = soup.select('.b_list.category > .bl_body li .bl_wrap > .bl_subject > a')           
    
    for title in titles:
        pet_title.append(title.get_text()[3:])
    
    for url in urls:
        pet_url.append('https://www1.president.go.kr' + url['href'])
     
data = {'제목':pet_title, 'URL':pet_url}
df = pd.DataFrame(data)


for n in tqdm_notebook(df.index): 
    url = urlopen(df['URL'][n]) 
    time.sleep(1)
    soup = BeautifulSoup(url, "html.parser")
    tmp = soup.select('h2.petitionsView_count > span.counter')
    for t in tmp:
        x = t.get_text()
        pet_ppl.append(x)
    tmp_x = soup.find('ul',class_='petitionsView_info_list').get_text()
    p0 = re.compile('\s')
    splited = p0.split(tmp_x) #tmp_y.li.get_text()
    category.append(splited[1:5][0][4:])
    pet_start.append(splited[1:5][1][4:])
    pet_end.append(splited[1:5][2][4:])
    tmp_y = soup.find_all('div',class_='View_write')
    if len(tmp_y) > 1:
        content_raw = tmp_y[1].get_text()
    else:
        content_raw = tmp_y[0].get_text()
    p1 = re.compile('\r|\t|\n|\xa0')
    content_splited = p1.split(content_raw)
    content.append(''.join(content_splited)) # 청원내용 가져오기
    
df['카테고리'] = category
df['청원내용'] = content
df['청원시작일'] = pet_start
df['청원종료일'] = pet_end
df['참여인원'] = pet_ppl


df.head()
```
