```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
import matplotlib as mpl
import matplotlib.pyplot as plt
import nltk # Natural Language ToolKit
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
text = open('Biden-victory-speech.txt','r',encoding='utf-8')
text1 = text.read()
#text1
#b_mask = np.array(Image.open('biden.png'))
text.close()
```

```python
stopwords = set(STOPWORDS)
stopwords.add("America")
stopwords.add("It")
stopwords.add("We will")
stopwords.add("will")
stopwords.add("ve")
stopwords.add("nation")
stopwords.add("American")
stopwords.add("President")
stopwords.add("president")
```

```python
wc = WordCloud(background_color='white', max_words=2000, collocations=False,
              stopwords = stopwords)
wc = wc.generate(text1)
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
