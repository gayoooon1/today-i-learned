```python
def corr_san(sodae, x): # 종사자수 대비 산업체 수
    file_path = '최종-'+ str(sodae) + '분류-산업별-데이터-'+ str(x) + '.csv'
    df = pd.read_csv(file_path, encoding = 'utf-8', index_col =0)
    sns.set_style('darkgrid')
    sns.lmplot(x='종사자수 대비 산업체 수', y='인구 대비 확진자 수', data=df,height=8)
    plt.title('Manufacturing')
    plt.xlabel("Number of industries compared to the number of employees")
    plt.ylabel("Number of infected persons relative to population")
    plt.rc('font', size=20)
    plt.show()
```

```python
corr_san('소',7) # 종교단체
```

```python
corr_san('대',5) # 보건업 및 사회복지 서비스
```

```python
corr_san('대',4) # 교육 서비스
```

```python
corr_san('대', 0) # 제조업
```

```python
def corr_ingu(sodae, x): # 인구 대비 산업체 종사자 수
    file_path = '최종-'+ str(sodae) + '분류-산업별-데이터-'+ str(x) + '.csv'
    df = pd.read_csv(file_path, encoding = 'utf-8', index_col =0)
    sns.set_style('darkgrid')
    sns.lmplot(x='인구 대비 산업체 종사자 수', y='인구 대비 확진자 수', data=df,height=8)
    plt.xlabel("Number of industrial workers relative to population")
    plt.ylabel("Number of infected persons relative to population")
    plt.title('Health industry and social welfare service')
    plt.show()
```

```python
corr_ingu('소', 0)
```

```python
corr_ingu('대',5)
```

```python
def corr_pop_san(sodae, x): # 인구 대비 산업체 수
    file_path = '최종-'+ str(sodae) + '분류-산업별-데이터-'+ str(x) + '.csv'
    df = pd.read_csv(file_path, encoding = 'utf-8', index_col =0)
    sns.set_style('darkgrid')
    sns.lmplot(x='인구 대비 산업체 수', y='인구 대비 확진자 수', data=df,height=8)
    plt.xlabel('Number of industries compared to population')
    plt.ylabel("Number of infected persons relative to population")
    plt.title('Associations and Other organizations  services')
    plt.show()
```

```python
corr_pop_san('소',4) # 콜센터 및 텔레마케팅 서비스업
```

```python
corr_pop_san('대',4) # 교육 서비스업
```

```python
corr_pop_san('대',7) # 협회 및 단체 기타 서비스업
```
