```julia
abstract type AbstractMember end
#Abstract types are nodes in the type hierarchy: they group types together
#Member

mutable struct Member <: AbstractMember
    name::String
    age::Int
end

name(p::AbstractMember) = p.name
age(p::AbstractMember) = p.age

function greet_m(x::AbstractMember)
    return name(x), age(x)
end

#Teacher
abstract type AbstractTeacher <: AbstractMember end

mutable struct Teacher <: AbstractTeacher
    name::String
    age::Int
    teaches::String
end

teaches(t::AbstractTeacher) = t.teaches



#Student
abstract type AbstractStudent <: AbstractMember end

mutable struct Student <: AbstractStudent
    name::String     
    age::Int
    id_no::Int 
end

id_no(s::AbstractStudent) = s.id_no


abstract type AbstractVisiting <: AbstractStudent end

mutable struct Visiting <: AbstractVisiting
    name::String     
    age::Int
    id_no::Int
    valid_thru::Int
end

valid_thru(v::AbstractVisiting) = v.valid_thru

function greet(x::AbstractMember)
    if typeof(x) <: AbstractVisiting
        return greet_m(x), id_no(x), valid_thru(x)
    elseif typeof(x) <: AbstractStudent
        return greet_m(x), id_no(x)
    #end
    elseif typeof(x) <: AbstractTeacher
        return greet_m(x), teaches(x)
    else 
        return greet_m(x)
    end
end

members = [
    Teacher("Harry", 55, "Programming Languages"),
    Teacher("Natasha", 61, "Forbidden Archeology"),
    Student("YK", 22, 2051),
    Student("SH", 35, 4968),
    Visiting("Alice", 24, 9595, 181225),
    Visiting("Vanessa", 22, 9598, 190208)
]

print("A few CAU members...\n")
for member in members
    print(greet(member))
    print("\n")
end

print("\n")

members_sorted_on_age = sort(members, by=m -> (m.age))
print("Members sorted on age...\n")
for member in members_sorted_on_age
     print(greet(member))
     print("\n")
end
print("\n")

members_sorted_on_name = sort(members, by=m-> lowercase(m.name), rev=true)
print("Members sorted in reverse on name...\n")
for member in members_sorted_on_name
    print(greet(member))
    print("\n")
end
```

## 코드 설명   

### type & function
abstract type AbstractMember end : 가장 포괄적인 Member 추상형 선언   
mutable struct Member <: AbstractMember : 이에 대한 struct 만들어서 이름, 나이 추가  

`name(p::AbstractMember) = p.name`, `age(p::AbstractMember) = p.age`
-> AbstactMember 형이면 이름과 나이를 말할 수 있음   
`function greet_m(x::AbstractMember)`
-> 위를 바탕으로 이름과 나이를 말하는 function 생성

#### Member에 종속되는 type
abstract type AbstractTeacher <: AbstractMember end : 선생님    
mutable struct Teacher <: AbstractTeacher   
`teaches(t::AbstractTeacher) = t.teaches` -> 선생님 형만 가르치는 과목명을 말할 수 있음   

abstract type AbstractStudent <: AbstractMember end : 학생   
mutable struct Student <: AbstractStudent   
`id_no(s::AbstractStudent) = s.id_no` -> 학생형은 학번을 말할 수 있음   

abstract type AbstractVisiting <: AbstractStudent end : 교환학생 (학생에 종속)   
mutable struct Visiting <: AbstractVisiting   
`valid_thru(v::AbstractVisiting) = v.valid_thru` -> 교환학생형은 유효기간을 말할 수 있음   

`function greet(x::AbstractMember)`   
* AbstractMember 형일 때는(greet_m) + 각자 Member에 종속되는 type중에 추가로 만들어진 기능을 더 호출할 수 있음
